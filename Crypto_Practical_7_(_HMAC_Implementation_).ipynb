```python
import hmac
import hashlib
key = "xxyyzzabc"
print('Key : ',key)
message = "KNGopal"
print("Message : ",message)
encodedkey = str.encode(key)
print("Encoded Key : ",encodedkey)
hmac1 = hmac.new(encodedkey,message.encode('UTF-8'),hashlib.sha1)
digestvalue1 = hmac1.digest()
print("Digest Value : ",digestvalue1)
digestvalue2 = hmac1.hexdigest()
print("Digest Value in Hex Dec : ",digestvalue2)
```
