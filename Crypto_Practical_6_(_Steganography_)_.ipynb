```python
import numpy as np
import cv2
import stepic
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
print("Encryption")
text = input("Data : ")
file = "/content/loc.png"
img = Image.open(file)
img_stegano = stepic.encode(img, text.encode())
img_stegano.save("2.png")
img = cv2.imread("/content/loc.png")
img = cv2.resize(img, (450, 300))
print("Secret Image is : ")
cv2_imshow(img)
```

```python
print("Decryption")
file ="2.png"
img = Image.open(file)
decoded = stepic.decode(img)
print(" Data is : "+ str(decoded))
```
